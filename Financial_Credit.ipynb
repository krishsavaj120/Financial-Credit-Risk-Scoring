```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('german_credit_data_updated.csv')
```

```python
df.head(10)
```

```python
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
```

```python
df.head(10)
```

```python
if 'Unnamed: 0' in df.columns: df.drop(columns=['Unnamed: 0'], inplace=True)
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
```

```python
df['saving_accounts'] = df['saving_accounts'].fillna('none/unknown')
df['checking_account'] = df['checking_account'].fillna('none/unknown')
```

```python
df.head(10)
```

```python
df['default_status'] = df['credit_risk'].apply(lambda x: 1 if x == 2 else 0)
```

```python
df.head(10)
```

```python
df['monthly_burden'] = df['credit_amount'] / df['duration']
```

```python
df.head(10)
```

```python
df['age_group'] = pd.cut(df['age'], bins=[18, 30, 45, 60, 100], 
                         labels=['Young Adult', 'Adult', 'Senior', 'Elderly'])
```

```python
df.head(10)
```

```python
def define_risk(row):
    if row['saving_accounts'] == 'little' and row['credit_amount'] > 5000:
        return 'Critical Risk'
    elif row['saving_accounts'] in ['rich', 'quite rich']:
        return 'Low Risk'
    else:
        return 'Standard'
```

```python
df['risk_tier'] = df.apply(define_risk, axis=1)
```

```python
df.head(10)
```

```python
df.to_csv('credit_risk_engineered.csv', index=False)
```
